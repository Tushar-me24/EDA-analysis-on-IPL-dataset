```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
ipl = pd.read_csv(r'C:\Users\Tushar\Downloads\matches.csv')
```

```python
#it returns the first 5 heads from the dataset
ipl.head()
```

```python
#shows the number of rows and columns in the dataset
ipl.shape
```

```python
#getting the frequency of most man of the match awards
ipl['player_of_match'].value_counts()
```

```python
#getting the top 10 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:10]
```

```python
#getting the top 5 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:5]
```

```python
#making bar-plot for the top 5 players with most man of the match awards
# Get top 5 players
top5_players = ipl['player_of_match'].value_counts()[0:5]

# Plotting
plt.figure(figsize=(6,5))
plt.bar(list(top5_players.keys()), list(top5_players.values), color="b")
plt.xlabel("Player")
plt.ylabel("Number of Awards")
plt.title("Top 5 Players with Most Man of the Match Awards")
plt.show()
```

```python
#getting the frequency of result column
ipl['result'].value_counts()
```

```python
#finding out the number of toss wins w.r.t. eash team
ipl['toss_winner'].value_counts()
```

```python
#extracting the records where a team won batting first
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
#looking at the head
batting_first.head()
```

```python
#making a histogram
plt.figure(figsize=(7,7))
plt.hist(batting_first['win_by_runs'],color="g")
plt.title('Distribution of runs')
plt.xlabel('runs')
plt.show()
```

```python
#finding out the number of wins w.r.t. each team after batting first
batting_first['winner'].value_counts()
```

```python
#making a bar plot for top 3 teams with most wins after batting first
plt.figure(figsize=(7,7))
plt.bar(
    list(batting_first['winner'].value_counts()[0:3].keys()),
    list(batting_first['winner'].value_counts()[0:3].values),
    color=["blue","yellow","orange"]
)
plt.xlabel("Team")
plt.ylabel("Wins After Batting First")
plt.title("Top 3 Teams with Most Wins After Batting First")
plt.show()
```

```python
#making a pie chart
plt.figure(figsize=(7,7))

wins = batting_first['winner'].value_counts()

plt.pie(
    list(wins),
    labels=list(wins.keys()),
    autopct='%1.1f%%'
)
plt.title("Wins by Teams After Batting First")
plt.show()
```

```python
#extracting those records where a team has won after batting second
batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
#looking at the head
batting_second.head()
```

```python
#making histogram for frequency of wins w.r.t. number of wickets
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'],bins=30,color="r")
plt.title('Distribution of runs')
plt.xlabel('runs')
plt.show()
```

```python
#finding out the frequency of number of wins w.r.t.each time after batting second
batting_second['winner'].value_counts()
```

```python
#making a bar plot for top 3 teams with most wins after batting second
plt.figure(figsize=(7,7))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color=['purple','blue','red'])
plt.show()
```

```python
#making a pie chart for distribution of most wins after batting second
plt.figure(figsize=(7,7))

wins = batting_second['winner'].value_counts()

plt.pie(
    list(wins),
    labels=list(wins.keys()),
    autopct='%1.1f%%'
)
plt.title("Wins by Teams After Batting second")
plt.show()
```
